```python
# 사용자 함수를 적용시키는 것이 중요 ⭐⭐⭐⭐⭐

import math

def avg_ceil(x,y,z):
  return math.ceil((x+y+z)/3) # 평균 계산 후 올림

abalone_df[['diameter','height','whole_weight']].apply(lambda x: avg_ceil(x[0],x[1],x[2]),axis = 1)
```

```python
# 문제
# 1. ['diameter','height','whole_weight'] 사용
# 2, 세 변수의 합이 1이 넘으면 True, 아니면 false 출력 후 answer 변수에 저장
# 3. ablone_df에 answer열을 추가하고 입력
```

```python
def sum_boolean(x):
  sum = x[0]+x[1]+x[2]
  if sum >= 1:
    return True
  return False

answer = abalone_df[['diameter','height','whole_weight']].apply(lambda x: sum_boolean(x),axis = 1)
abalone_df['answer'] = answer
#new_abalone_df = pd.concat([abalone_df,answer],axis = 1)
abalone_df
```
