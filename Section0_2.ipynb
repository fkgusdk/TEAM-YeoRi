# DataFrame 행,열 선택 및 필터링

```python
import pandas as pd

# 베이스 경로
base_src = './drive/MyDrive/ㅇ3ㄹ1/'

# 데이터 경로 
friend_src = base_src+"/friend.csv"

# 불러오기
df = pd.read_csv(friend_src,encoding='utf-8')

df.head()
```

# row 가져오기

```python
# index 2번에 해당하는 row 가져오기 (series 형태)
df.iloc[2]
```

```python
# index 2번에 해당하는 row 가져오기 (dataframe 형태)
df.iloc[[2]]
```

```python
# 열 가져오기 (series)
df['job']
```

```python
df.loc[:'job']
```

```python
# 열 여러개 가져오기 (dataframe)
df[['job','name']]

# df['job','name']은 안 됨
```

# 참고

```python
# 참고
print(df.iloc[:2],'\n\n') # col 부분 생략가능

# print(df.iloc[,:2]) -> error (row 부분 생략 불가능)
print(df.iloc[:,:2])
```

```python
# 참고
# 열의 dtype은 문자 -> 실수 -> 정수 순서로 우선 순위이며,
# 가장 높은 우선 순위로 표시되기는 하지만 데이터 하나하나의 타입은 변하지 x

# 새로운 열 생성
df['Score'] = [99.1,78.9,88.5,65.0,'93.2',89]

# 확인
print(df['Score'],'\n\n')
print("df.loc[4,'Score'] : ",type(df.loc[4,'Score']))
print("df.loc[5,'Score'] : ",type(df.loc[5,'Score']))
print("df.loc[6,'Score'] : ",type(df.loc[6,'Score']))
```

```python
# 참고
# float가 많으면 int로 적어도 float
df['Score'] = [99.1,78.9,88.5,65.0,93.2,89]

# 확인
print(df['Score'],'\n\n')
print("df.loc[5,'Score'] : ",type(df.loc[5,'Score']))
print("df.loc[6,'Score'] : ",type(df.loc[6,'Score']))
```

# DataFrame 행,열 삭제

```python
import numpy as np
```

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
df.drop(['B','C'],axis=1) # inplce = True 하면 변수에 또 할당해주지 않아도 알아서 할당함!
# axis = 1 : 세로 축
```

```python
df.drop([0,1])
# axis = 0 : 가로 축 (default, 그래도 써주는게 좋음)
```
