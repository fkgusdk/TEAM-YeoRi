## Pie Chart

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 시각화 테이블 (?) 크기 
plt.rcParams['figure.figsize'] = [10,8] # (가로, 세로)

# 그 외 스타일들
sns.set(style='whitegrid') # 흰색 바탕 + 격자
sns.set_palette('pastel')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# matplotlib을 활용한 시각화
sum_tip_by_day = tips.groupby('day')['tip'].sum()
sum_tip_by_day
```

```python
### 참고
sum_tip_by_day2 = tips['tip'].groupby(tips['day']).sum()
# tips['tip'].groupby('day').sum() -> 오류
sum_tip_by_day2
```

```python
ratio_tip_by_day = sum_tip_by_day / sum_tip_by_day.sum()
ratio_tip_by_day 
```

```python
x_label = ['Thur','Fri','Sat','Sun']

plt.pie(ratio_tip_by_day,
        labels=x_label,
        autopct='%.1f%%', # 파이 내부에 정확한 % 기입해 줌 (소수점 한자리까지 표시)
        )

plt.show()
```

```python
# startangle ,counterclock
x_label = ['Thur','Fri','Sat','Sun']

plt.pie(ratio_tip_by_day,
        labels=x_label,
        autopct='%.1f%%', # 파이 내부에 정확한 % 기입해 줌 (소수점 한자리까지 표시)
        startangle=90, ### 파이 시작 각도 (축이 시작되는 각도)
        counterclock=True, # 반시계 방향으로 라벨 순서대로 나열 (False면 시계, 없으면 반시계방향으로 파이크기 순서대로)
        explode=[0.25,0.05,0.05,0.05], # 파이 사이 간격
        shadow=True,
        colors=['gold','silver','whitesmoke','green'],
        wedgeprops={'width':0.7,'edgecolor':'r','linewidth':3} # 도넛모양 만들기 (안쪽 빈공간의 넓이, 색, 둘레 선 두께)
        )

plt.show()
```
