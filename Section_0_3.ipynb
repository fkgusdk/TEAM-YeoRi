## 조건 필터링 가져오기

```python
# 30대 이상만 가져오기
df[df.age >= 30]
```

```python
# job이 인턴인 사람 가져오기
df[df.job == 'intern']
```

```python
# 조건 여러게
# 조건 괄호로 묶기!
df[(df.age >= 30) & (df.age < 45)]
```

```python
df[(df.age < 30) | (df.age > 40)]
```

```python
# in 을 통한 포함 조건 걸기
df[df['name'].apply(lambda x: x in ['Jenny', 'Nate'])]
```
